```python
df = pd.read_excel("elTab.xlsx")
df.head()
```

```python
addresses = []
bar_address_dict = {}

for index,row in df.iterrows():
    
    bar_name = row[0].replace(' ', '+')
    location_name = str(row[1]).replace(' ','+')

    if pd.isnull(row[1]) == True:
        search_term = bar_name +'+'+ 'London+Address'
    else:
        search_term = bar_name +'+'+ location_name +'+'+ 'London+Address'
    
    url = f"https://www.google.com.tr/search?q={search_term}"
    
    print('Searching address for: ', bar_name)
    
    try:
        # Get Google page + parse
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'html.parser')

        with requests.Session() as s:
            
            headers = {
                "referer":"referer: https://www.google.com/",
                "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36"
                }
            s.post(url, headers=headers)
            response = s.get(url, headers=headers)
            soup = BeautifulSoup(response.text, 'html.parser')
            result = soup.find_all("div", {"class": "sXLaOe"})
            
            if result == []:
                print('no address found')
            for res in result:
                address = res.decode_contents().strip()
                addresses.append(address)
                print('Address found for ', row[0], ' is ', address )
                bar_address_dict[row[0]] = address
    except:
        print("Failed searching for address of:", bar_name)
        addresses.append('')
        bar_address_dict[row[0]] = ''
    time.sleep(5)
```
